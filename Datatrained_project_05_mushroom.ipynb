```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

#### EDA

```python
mapping = list()

encoder = LabelEncoder()

for column in range(len(df.columns)):
    df[df.columns[column]] = encoder.fit_transform(df[df.columns[column]])
    mappings_dict={index:label for index,label in enumerate(encoder.classes_)}
    mapping.append(mappings_dict)
```

```python
x = df.drop(['class'],axis=1) 
y = df['class']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x = pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test= train_test_split(x,y,test_size =.2)
```

#### Model selection

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
from sklearn.svm import SVC
svc = SVC(C =1,kernel = 'rbf')
```

#### Training

```python
np.sum(y)/len(y)    
```

The above data means our data is 48% positive and 52% negative. its a quite balanced data

```python
lr.fit(x_train,y_train)
svc.fit(x_train,y_train)
```

```python
print(f'logistic regression: {lr.score(x_test,y_test)}')
print(f'Support Vector Machine: {svc.score(x_test,y_test)}')
```

## Support vector machine is the best fitting model it is giving 100 percent accuracy. So this is a clean data
